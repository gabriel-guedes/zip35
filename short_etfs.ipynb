```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Function for recalculating series with common base
def rebased(ts):
    return ts / ts[0]
```

```python
# Read data from csv
df = pd.read_csv('./data/short_etfs.csv', index_col='Date', parse_dates=['Date'])
df.head()
```

```python
# Calculate all series starting from first value
df = df.apply(rebased)
df.head()
```

```python
# Make new figure and set the size
fig = plt.figure(figsize=(12,8))

# The first subplot, planning for 3 plots high, 1 plot wide, this being the first 
ax = fig.add_subplot(111)
ax.set_title('Short ETF effect')
ax.plot(df['SPY'], label='SPY index tracker', linewidth=3.0, color='black')
ax.plot(df['SH'], label='SH Short ETF', linewidth=3.0, color='grey')
ax.plot(df['SDS'], label='SDS Double Inverse ETF', linewidth=2.0, color='grey')
ax.legend()
```
